## 1. Factorial of a Number

Write a recursive program to find the factorial of a number.

```
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)

num = 5
print("Factorial:", factorial(num))
```

## 2. Fibonacci Number (Nth Term)

Write a recursive program to find the nth Fibonacci number.

```
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

n = 6
print("Fibonacci term:", fibonacci(n))
```

## 3. Sum of Digits of a Number

Write a recursive program to find the sum of digits.

```
def sum_digits(n):
    if n == 0:
        return 0
    return n % 10 + sum_digits(n // 10)

num = 1234
print("Sum of digits:", sum_digits(num))
```

## 4. Reverse a Number

Write a recursive program to reverse a number.

```
def reverse_number(n, rev=0):
    if n == 0:
        return rev
    return reverse_number(n//10, rev*10 + n%10)

num = 1234
print("Reversed number:", reverse_number(num))
```

## 5. Check Palindrome Number

Write a recursive program to check whether a number is palindrome.

```
def reverse_number(n, rev=0):
    if n == 0:
        return rev
    return reverse_number(n//10, rev*10 + n%10)

num = 121

if num == reverse_number(num):
    print("Palindrome number")
else:
    print("Not a palindrome")
```

## 6. Power of a Number (xⁿ)

Write a recursive program to find power of a number.

```
def power(x, n):
    if n == 0:
        return 1
    return x * power(x, n-1)

print("Power:", power(2, 5))
```

## 7. GCD of Two Numbers

Write a recursive program to find GCD using Euclid’s algorithm.

```
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

print("GCD:", gcd(48, 18))
```

## 8. Print Array Elements Using Recursion

Write a recursive program to print array elements.

```
def print_array(arr, index=0):
    if index == len(arr):
        return
    print(arr[index])
    print_array(arr, index+1)

arr = [1, 2, 3, 4, 5]
print_array(arr)
```

## 9. Find Maximum Element in an Array

Write a recursive program to find the maximum element.

```
def find_max(arr, n):
    if n == 1:
        return arr[0]
    return max(arr[n-1], find_max(arr, n-1))

arr = [3, 9, 2, 5, 7]
print("Maximum element:", find_max(arr, len(arr)))
```

## 10. Tower of Hanoi

Write a recursive program to solve Tower of Hanoi.

```
def tower_of_hanoi(n, source, auxiliary, destination):
    if n == 1:
        print(f"Move disk 1 from {source} to {destination}")
        return
    tower_of_hanoi(n-1, source, destination, auxiliary)
    print(f"Move disk {n} from {source} to {destination}")
    tower_of_hanoi(n-1, auxiliary, source, destination)

tower_of_hanoi(3, 'A', 'B', 'C')
```
