## Program 8: Prime Number Checker 
Problem: Check if a number is prime. Use break when divisor found. 
 
number = int(input("Enter a number: ")) 
### Your task:  
### - Check divisibility from 2 to number-1 
### - Use break when divisor found 
### - Print if prime or not

```python
number = int(input("Enter a number: ")) 

ref = True
for i in range(2,number):
    
    if number % i == 0:
        
        ref = False
        break
    
if ref:
    print("prime number ")
    
else:
    
    print("Not Prime Number")
```
