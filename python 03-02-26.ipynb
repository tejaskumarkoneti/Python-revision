## upper() – Convert to uppercase 

```python
text = "python" 
```

```python
print(text.upper())
```

## Convert to uppercase - without built in

```python
res = ""

text ="python"

for ch in text:
    
    if ord(ch) >= ord("a") and ord(ch) <= ord("z"):
        
        res_int = ord(ch) - 32
        
        res_ch = chr(res_int)
        
        res = res + res_ch

print(res)
```

## lower() – Convert to lowercase 

```python
email = "NAGA@GMAIL.COM" 
email.lower() 
```

## Convert to lowercase without built in (not expected output)

```python
email = "NAGA@GMAIL.COM" 

res = ""

for ch in email:
    
        
    if ord(ch) >= ord("A") and ord(ch) <= ord("Z"):
        
        res_int = ord(ch) + 32
        
        res_ch = chr(res_int)
        
        res += res_ch
    
    else: 
    
        res += ch

print(res)
```

###  title() – First letter of every word capital 
##### name = "naga balla" 
##### name.title() 
#####   Output: "Naga Balla" 
#####    Use case: Names, headings

```python
name = "naga balla"

res1 = ""
res1 += name[0]
for ch in range(len(name)):
    
    if ch == " ":
        res1 += name[ch+1]
        break
    
print(res1)
    
    
```

     Trimming Functions (Very Common) 
 strip() – Remove spaces (both sides) 
name = "  Naga  " 
name.strip() 
   Output: "Naga" 
    Use case: User input cleaning 
 
 lstrip() – Remove left spaces 
text = "   Hello" 
text.lstrip() 
   Output: "Hello" 
 
 rstrip() – Remove right spaces 
text = "Hello   " 
text.rstrip() 
   Output: "Hello" 
 
    Replace & Modify (Creates NEW string) 
 replace(old, new) 
msg = "I like Java" 
msg.replace("Java", "Python") 
   Output: "I like Python" 
    Use case: Updating text, masking words 
 
 split() – String → List 
text = "Python is easy" 
text.split() 
   Output: ["Python", "is", "easy"] 
    Use case: Breaking sentences, CSV data 
 
 join() – List → String 
words = ["Python", "is", "easy"] 
" ".join(words) 
   Output: "Python is easy" 
     Logic to remember: 
• split() → breaks 
• join() → connects
