##  Count how many times a word appears in a sentence (without In Built Functions)

```python
sentence = "Yesterday a turtle lived in India went to laundry found a secret key"

search_word = input("Enter the word to search in sentence ")

st = ""

li = []

for ch in sentence:
    
    if ch == " ":
        
        li += [st]
        st =""
        
    else:
        
        st += ch
        

count = 0
        
for item in li:
    
    if item == search_word:
        
        count +=1
        
print(count)
```

##  Find the largest number entered until the user types “stop” 

```python
larg = 0

while True:
    
    user = input("Enter the number (or Type Stop) : ")
    if user == "stop":    
        break
    val = int(user)
    if larg == 0 or val > larg:
        larg = val
        
print("Largest number is {} ".format(larg))
```

### 2. Count the number of vowels and consonants in a string. 


```python
st = "Rajesh"

vowels = ['a','e','i','o','u']

v_count = 0

c_count = 0

for ch in st:
    
    if ch in vowels:
        
        v_count +=1
        
    else:
        
        c_count +=1
        
print("No of vowels in {} are {}".format(st,v_count))

print("No of constants in {} are {}".format(st,c_count))
```

### 3. Reverse a string. 


```python
st = "Rajesh"

for char in range(len(st)-1,-1,-1):
    
    print(st[char],end="") 
```

### 4. Check whether a string is a palindrome. 


```python
st1 = "mada"


rev = ""

for ch in range(len(st1)-1,-1,-1):
    
    rev = rev + st1[ch]

print(rev)
    
if rev == st1:
    
    print("Palindrome")

else:
    
    print("Not Palindrome")
```

### 5. Count uppercase, lowercase, digits, and special characters. 


```python
upper = 0
lower = 0
digit = 0
special = 0

st1 ="Hello@123"

for ch in st1:
    
    if "A" <= ch <= "Z":
        upper += 1
    elif 'a' <= ch <= 'z':
        lower += 1
    elif '0' <= ch <= '9':
        digit += 1
    else:
        special += 1

print(upper,lower,digit,special)
```

### 7. Replace spaces with underscores. 


```python
st = "hello world how are you"

res = ""

for ch in st:
    
    if ch == " ":
        
        res = res + "_"
    
    else:
        
        res = res + ch

print(res)
```

### 8. Remove duplicate characters from a string. 


```python
st1 = "banana"

res = ""

for ch in st1:
    
    if ch in res:
        
        continue
    
    else:
        
        res += ch

print(res)
```

```python
stg = "madam"

lt = []

for index in range(len(stg)):
    
    ch = stg[index]
    
    if ch != " ":
        
        lt += [ch]
    
print(lt)     

res = ""

for item in lt:
    
    if item not in res:
        
        res += item
        
print(res)   
        
```

### 10. Find the first non-repeating character. 


```python
st1 = "aabbcde"

my_dict = {}

for ch in st1:
    k = ch
    if k in my_dict:
        my_dict[k] +=1
    else:
        my_dict[k] = 1

print(my_dict)

for k in st1:
    print(k)
    if my_dict[k] == 1:
        print("The first non-repeating character is {}".format(k))
        break
```

# INTERMEDIATE STRING PROBLEMS (Logic Building) 


### 11. Check whether two strings are anagrams. 


```python
def anagram(s):
    d = {}
    for i in s:
        if i in d:
            d[i] += 1
        else:
            d[i] = 1
    return d

st1 = input("Enter string 1 : ")
st2 = input("Enter string 2 : ")

check_len = len(st1) == len(st2)

if check_len:
    a1 = anagram(st1)
    a2 = anagram(st2)
    print(a1)
    print(a2)
    if a1 == a2:
        print("Anagram")
    else:
        print("Not Anagram")    
```

### 12. Find the most frequent character in a string. 


```python
str1 = "banaba"
d = {}
for ch in str1:
    if ch in d:
        d[ch] += 1
    else:
        d[ch] = 1
print(d)

max_count = 0
max_char = None
for k,v in d.items():
    if v > max_count:
        max_count = v
        max_char = k
print(max_char)
```
