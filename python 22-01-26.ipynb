## Print bill amount for items in a cart 

```python
amount = 0
cart = {'soap': 45,'tshirt':200,'choclate':20}
for iter in cart.values():
    amount += iter
print(amount)
```

## Print table of a number

```python
num = 9
for i in range(1,11):
    print("{} * {} = {}".format(num,i,num*i))
```

```python
#updated code
num = 10
for i in range(1,11):
    print(f"{num} * {i} = {num * i}")
```

## Count how many times a word appears in a sentence

```python
sentence = "Yesterday a turtle lived in India, went to laundry and found a secret key"
s = sentence.split()
print(s)
count = 0
for word in s:
    if word == "a":
        count +=1
print(count)
```

##  Manual splitting


```python
stmt = "learning python is simple python is nice"
# Manual splitting

d = []             # we have taken empty list
st = ""            # intialize the string 
for ch in stmt:    # iterate over stmt string
    if ch == " ":  # if ch is space 
        d = d + [st]  # add the entire string to the list 
        st = ""       # intialize the string to empty
    else:
        st = st + ch  # if ch is not space then add the characters so that if forms a  string 
        
d = d + [st]         # add the last word ('nice') in to list  
print(d)
```
