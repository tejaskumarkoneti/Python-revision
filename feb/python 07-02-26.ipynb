```python
d = {"x":10,
     "y":20,
     "z":30 }

for i in d:
    print(i, "=",d[i])
```

```python
for i in d.keys():
    
    print(i)
```

```python
for i in d.values():
    
    print(i)
```

```python
for k,v in d.items():
    
    print(k,"=",v)
```
