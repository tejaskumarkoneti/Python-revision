# 17. Find the longest word in a sentence. 

#### Input  : "Python is a powerful programming language"

#### Input  : "I love machine learning"
#### Output : "learning"  (8 characters)
#### Input  : "Hi"

```python
sentence = "Python is a powerful programming language"
sentence = "aa b"

words = sentence.split()           # Step 1: Break into list of words

longest = ""                       # Step 2: Start with empty word

for word in words:                 # Step 3: Check each word
    if len(word) > len(longest):   # Step 4: Is this word longer than longest?
        longest = word             # Step 5: If yes, update longest

print("Longest word :", longest)
print("Its length   :", len(longest))

#Output:
#Longest word : programming
#Its length   : 11
```

### 20. Compress a string (example: aaabbc → a3b2c1). 
Input:

aaabbc


Output:

a3b2c1


```python
s = "aaabbc"
d = {}

for ch in s:
    
    if ch in d:
        
        d[ch] += 1
        
    else:
        
        d[ch] = 1

print(d)

res = ""

for k,v in d.items():

    res = res +  k + str(v)
    
print(res)
```

### • Check password strength. 

```python
def check_password_strength(s):

    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False

    for ch in s:
        if ch.islower():
            has_lower = True
        if ch.isupper():
            has_upper = True
        if ch.isdigit():
            has_digit = True
        if not ch.isalnum():
            has_special = True
    type_count = has_lower + has_upper + has_digit + has_special

    if len(s) < 6 or type_count == 1:
        return "Weak"
    elif type_count == 2:
        return "Medium"
    else:
        return "Strong"
    
# Example
print(check_password_strength("AAAAb@1234"))  # Strong
```

```python
url = "https://www.google.com/search"

# Step 1: Remove protocol (http:// or https://)
if "//" in url:
    url = url.split("//")[1]
print(url)

# Step 2: Remove path (anything after '/')
url = url.split("/")[0]

# Step 3: Remove 'www.' if present
if url.startswith("www."):
    url = url[4:]
print(url)
```
