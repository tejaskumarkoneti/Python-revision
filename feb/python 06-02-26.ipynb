Example 1

Input:
"482915"

Output:
Valid OTP (Only digits)

Example 2

Input:
"98A731"

Output:
Invalid OTP (Contains non-digit characters)


```python
"482915".isdigit()
```

```python
"48291A".isdigit()
```

```python
user = input("Enter the OTP :")

if user.isdigit():
    
    print("True")

else:
    
    print("False")
```

```python
li = list(range(0,9))

user = input("Enter the OTP :")  # 482915

ref = None

for ch in user:
    
    if ch in li:
        
        ref = True
    
    ref = False
        

if ref:
    print("Valid OTP")
else:
    print("Invalid OTP")
```
