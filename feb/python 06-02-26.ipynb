Example 1

Input:

Valid OTP (Only digits)

Example 2

Input:
"98A731"

Output:
Invalid OTP (Contains non-digit characters)


```python
"482915".isdigit()
```

```python
"48291A".isdigit()
```

```python
user = input("Enter the OTP :")

if user.isdigit():
    
    print("True")

else:
    
    print("False")
```

```python
li = list(range(0,9))

user = input("Enter the OTP :")  # 482915

ref = None

for ch in user:
    
    if ch in li:
        
        ref = True
    
    ref = False
        

if ref:
    print("Valid OTP")
else:
    print("Invalid OTP")
```

### 16. Reverse each word in a sentence. 


```python
sentence = "The sunset is so beautiful in Agra"

sentence_li = sentence.split()

st = ""

for ele in sentence_li:
    
    st += ele[::-1]  + " "
    
print(st)
```

# 17. Find the longest word in a sentence. 

#### Input  : "Python is a powerful programming language"
#### Output : "programming"  (11 characters)
#### Input  : "I love machine learning"
#### Output : "learning"  (8 characters)
#### Input  : "Hi"
#### Output : "Hi"  (only one word)

```python
sentence = "Python is a powerful programming language"

words = sentence.split()           # Step 1: Break into list of words

longest = ""                       # Step 2: Start with empty word

for word in words:                 # Step 3: Check each word
    if len(word) > len(longest):   # Step 4: Is this word longer?
        longest = word             # Step 5: If yes, update longest

print("Longest word :", longest)
print("Its length   :", len(longest))

#Output:
#Longest word : programming
#Its length   : 11
```
