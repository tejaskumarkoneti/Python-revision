Count number of characters in a string 

```python
count = 0
st=input("Enter the string")
for ch in st:
    count+=1
print(count)
```

Count vowels in a string 

```python
count = 0
li= []
st1=input("enter the string :")
vowel = ['a','e','i','o','u']
for i in st1:
    if i in vowel:
        count+=1
        li += [i]
print(li)
print(count)
```

Reverse a string using for loop

```python
st1 = "rajesh"
for i  in range(len(st1)-1,-1,-1):
    print(st1[i],end=" ")
```

Print characters at even positions 

```python
stg1 = "manohar"
for i in range(len(stg1)):
    if i % 2 == 0:
        print(stg1[i],end=" ")
```
