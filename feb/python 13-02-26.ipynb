### 1. Check if a Number is Even or Odd

**Question:**  
Given an integer N, determine whether it is even or odd.

**Sample Input:**  
7  

**Sample Output:**  
Odd  

**Explanation:**  
Any number divisible by 2 is even, otherwise odd.


### 2. Find the Largest of 3 Numbers

**Question:**  
Given three integers A, B, and C, print the largest number.

**Sample Input:**  
5 12 9  

**Sample Output:**  
12  

**Explanation:**  
Compare all three values and print the maximum.


### 3. Reverse a Number

**Question:**  
Given an integer, reverse its digits.

**Sample Input:**  
1234  

**Sample Output:**  
4321  

**Explanation:**  
Digits are read from right to left.


### 4. Sum of Digits

**Question:**

**Sample Input:**  
538

16  

**Explanation:**  
5 + 3 + 8 = 16


### 5. Palindrome Number Check

**Question:**  
Check whether a number reads the same forwards and backwards.

**Sample Input:**  
121  

**Sample Output:**  
Palindrome  

**Explanation:**  
121 reversed is still 121.

### 7. Factorial of a Number

**Question:**  
Calculate factorial of N.

**Sample Input:**  
5  

**Sample Output:**  
120  

**Explanation:**  
5! = 5×4×3×2×1
