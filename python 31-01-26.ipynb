### Challenge 1: Menu-Driven Calculator 

### Create calculator with menu: 
##### 1. Add 
##### 2. Subtract 
##### 3. Multiply 
##### 4. Divide 
##### 5. Exit 
### Use break to exit, continue for invalid choices

```python
print("Menu Driven Calculator")

print("\tMENU ITEMS are ")
print("1.ADD")
print("2.SUBRACT")
print("3.MULTIPLY")
print("4.DIVIDE")
print("5.EXIT")

min = 1
max = 6
for choice in range(min,max+1):
    
    user_choice = int(input("Enter (1- 5) from choices :"))
    
    if user_choice == 1:
        print("Addition",min+max)
    elif user_choice == 2:
        print("Subraction",max-min)
    elif user_choice == 3:
        print("Multiplication",min * max)
    elif user_choice == 4:
        print("Divison",max/min)
    elif user_choice == 5:
        print("User Wants Exit , So Exiting")
        break
    else:
        print("Your choice is Invalid")
        continue
```

# Mini Project Description â€“ Number Guessing Game

## Aim
To develop a simple **Number Guessing Game** that helps beginners understand basic programming concepts such as:
- User input  
- Conditional statements  
- Loops  
- The use of the `continue` statement  

---

## Project Description
In this mini project, the computer randomly selects a number within a given range.  
The user must guess the number by entering inputs repeatedly until the correct number is guessed.

If the user enters an **invalid input** (for example, a number outside the allowed range), that attempt is skipped and the program continues asking for the next guess.

This is where the **`continue` statement** is applied.

---

## What the User Needs to Do
- The user is asked to guess a number.
- If the guess is wrong, the program asks again.
- If the guess is outside the allowed range, that attempt is ignored and the game continues.
- If the guess is correct, the game ends with a success message.

---

## Key Concepts Used
- User input  
- Conditional statements  
- Looping  
- `continue` statement  
- Basic logic building  

---

## Learning Outcome
By completing this mini project, students will learn how to:
- Control the flow of a loop  
- Skip invalid cases using `continue`  
- Build simple game logic  
- Improve problem-solving skills  

---

## Real-Life Connection
Just like **skipping ads on YouTube and continuing to watch videos**, the program skips invalid guesses and continues the game until the correct number is found.
