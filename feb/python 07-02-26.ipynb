## 3. Separate Even and Odd Numbers into Two Lists

Write a program to separate even and odd numbers into different lists.

```python
lis = [1,67,107,2,8,4]

idx = 0
o_list = []
e_list = []

while idx < len(lis):
    if lis[idx] % 2 == 0:
        e_list += [lis[idx]]
    else:
        o_list += [lis[idx]]
    idx +=1

print(e_list)
print(o_list)
```

## 4. Check Whether a List is a Palindrome

Write a program to check if a list reads the same forward and backward.

```python
#p_li = ['r','a','j','e','s','h']
p_li = ['r','a','c','e','c','a','r']

for idx in range(len(p_li)):
    
    if p_li[idx] == p_li[len(p_li)-idx-1]:
        print(f"{p_li} is a palindrome")
    else:
        print("Not a palindrome")
```

## 6. Move All Zeros to the End of the List

Write a program to move all zeros to the end while maintaining order.                                                                
Example 1

Input:
[0, 1, 0, 3, 12]

Output:
[1, 3, 12, 0, 0]

Example 2

Input:
[4, 0, 5, 0, 0, 3, 1]

Output:
[4, 5, 3, 1, 0, 0, 0]

```python
z_n_nz = [4, 0, 5, 0, 0, 3, 1]

new_nz_z = []

L = len(z_n_nz)

for idx in range(L):
    
    if  z_n_nz[idx] != 0:
        
        new_nz_z += [z_n_nz[idx]]

for idx in range(L- len(new_nz_z)):
    
    new_nz_z += [0]

print(new_nz_z)
```

## 7. Find Common Elements Between Two Lists

Write a program to find common elements in two lists.
Example 1

Example 2

Example 3

Input:
List 1: [2, 4, 6, 8]
List 2: [1, 2, 3, 4, 5]

Output:
[2, 4]

```python
li1 =  [1, 2, 3, 4, 5]
li2 = [3, 4, 5, 6, 7]

res = []

for idx1 in range(len(li1)):
    
    for idx2 in range(len(li2)):
        
        if li1[idx1] == li2[idx2]:
            
            res += [li1[idx1]]
            
print(res)
```

## 9. Find the Element with the Highest Frequency

Write a program to find the most frequent element.

Example 1

Input:
[1, 3, 2, 3, 4, 3, 5]

Output:
3

Example 2

Input:
[7, 7, 8, 9, 8, 7, 8, 8]

Output:
8

Example 3

Input:
[10, 20, 10, 30, 20, 10]

Output:
10

```python
lihf = [1, 3, 2, 3, 4, 3, 5]

dhf = {}

for i in range(len(lihf)):
    
    k = lihf[i]
    
    if k in dhf:
        dhf[k] += 1
    
    else:
        dhf[k] = 1

print(dhf)

lar = dhf[1]

for k in dhf:
    
    if dhf[k]  > lar:
        
        lar = dhf[k]

print(lar)
```

## 10. Sort a List Without Using sort() Method

Write a program to sort a list using Bubble Sort.

```python
li = [5, 2, 9, 1, 7]

for i in range(len(li) - 1):
    
    if li[i] > li[i+1]:
        
        li[i] , li[i+1] = li[i+1],li[i]
        
print(li)
        
```
