# Print Fibonacci series up to n terms

```python
x = 0
y = 1
n = 7
while x < n:
    print(x)
    x = y
    y = x + y
```

# Find count of digits in a number

# Find sum of digits in a number

```python
sum = 0
n = 1892
while n > 1:
    n = n % 10
    sum += n
    n = n //10
print(sum)
```

# Find sum of digits in a number (updated)

```python
sum = 0
n = 1892
for i in range(n):
    n = n % 10
    sum += n
print(sum)
```
