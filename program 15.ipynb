```python
# Manual Splitting

sentence = "Python is an amazing programming language"

li = []

for ch in sentence:
    st=""               # in each interation the string is reset to empty string
    if ch == " ":
        li = li + [st]
        st =""
    else:                   # if ch contains alphabets then it adds each alphabet to the string
        st = st + str(ch)   # the string concatination
print(li)
```
