```python
def prime(n):
    if n < 2:
        return False
    i = 2
    while i < n:
        if n % i == 0:   # divisible → not prime
            return False
        i += 1
    return True           # no divisor found → prime


count = 2
number = 25
while count <= number:    # loop from 2 up to 25
    if prime(count):      
        print(count, end=" ")
    count += 1
```

```python
def fib(n):
    a = 0
    b = 1
    while a < n:
        print(a)
        a,b = b,a+b

number = int(input("enter number :"))
fib(number)
```
