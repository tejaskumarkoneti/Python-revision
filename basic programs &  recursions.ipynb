```python
import math

def countDigits(n):
    if n == 0:
        return 1  # 0 has 1 digit
    # formula: floor(log10(|n|)) + 1
    return int(math.log10(abs(n))) + 1


# Example usage
print(countDigits(43332))  # 5
print(countDigits(1000))   # 4
print(countDigits(7))      # 1
print(countDigits(0))      # 1
print(countDigits(-9876))  # 4
```

```python
def prime(n):
    import math
    if n < 2:
        return False
    else:
        for i in range(2, int(math.sqrt(n)+1)):   # loop only till n-1
            if n % i == 0:
                return False    # found a divisor → not prime
        return True             # no divisors found → prime


# Example usage
n = 29
if prime(n):
    print(n, "is a prime number")
else:
    print(n, "is not a prime number")
```

```python
def is_prime(n):
    if n < 2:  # 0 and 1 are not prime
        return False
    for i in range(2, int(n**0.5) + 1):  # check divisibility up to sqrt(n)
        if n % i == 0:
            return False
    return True

# Print primes below 100
for num in range(2, 100):
    if is_prime(num):
        print(num, end=" ")
```

```python
def reverseArray2(arr,n):
    p1 = 0
    p2 = n -1
    while p1 < n/2:
        arr[p1], arr[p2] = arr[p2],arr[p1]
        p1 +=1
        p2 -=1
    return arr

arr = [5, 4, 3, 2, 1]
n = len(arr)

p =  reverseArray2(arr, n)
print(p)
```

```python
def reverseArray(arr, start, end):
    if start < end:
        arr[start], arr[end] = arr[end], arr[start]
        reverseArray(arr, start + 1, end - 1)
        return arr
arr = [1,2,3,4,5]
n = len(arr) 
reverseArray(arr, 0,len(arr)-1)
```

```python
def fibonacci(N):
    # Base condition
    if N <= 1:
        return N
    
    # Recursive calls for previous two terms
    last = fibonacci(N - 1)
    slast = fibonacci(N - 2)
    
    return last + slast

# Main part
N = 4  # example value
print(fibonacci(N))
```

```python
s1 = "A man, a plan, a canal: Panama"

# Step 1: Filter alphanumerics and lower case
s2 = ""
for ch in s1:
    if ch.isalnum():
        s2 += ch.lower()

print(s2)       # amanaplanacanalpanama
print(len(s2))  # length of filtered string

# Step 2: Reverse using two-pointer approach
chars = list(s2)          # convert string to list (mutable)
l = 0
r = len(chars) - 1

while l < r:
    chars[l], chars[r] = chars[r], chars[l]  # swap
    l += 1
    r -= 1

rev = "".join(chars)      # back to string
print(rev)
```

```python
def isPalindrome(s):
    left = 0
    right = len(s)-1
    while left < right:
        if not s[left].isalnum():
            left += 1
        elif not s[right].isalnum():
            right -= 1
        elif s[left].lower() != s[right].lower():
            return False
        else:
            left += 1
            right -= 1
    return True



if __name__ == "__main__":
    str = "A?BCDC?BA"
    ans = isPalindrome(str)


    if ans == True:
        print("Palindrome")
    else:
        print("Not Palindrome")
```

```python
def is_palindrome(s):
    n = len(s)
    for i in range(n // 2):  # loop till half the string
        if s[i] != s[n - i - 1]:  # compare first and last moving inward
            return False  # mismatch → not a palindrome
    return True  # completed loop → palindrome

# Example usage
s = "madam"
if is_palindrome(s):
    print("Palindrome")
else:
    print("Not Palindrome")
```
