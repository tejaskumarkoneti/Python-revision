Check whether a number is prime (updated version)

```python
def prime(n): # function defination
    if n <= 1:
        return True
    
    for i in range(2,n):
        
        if n % i == 0:  # divide the n from 2 to n-1 to check if n has any divisble no's
            return False
        
    return True

n = 12
is_Prime = prime(n)

if is_Prime:
    print("Prime number")
else:
    print("Not prime number")
```
