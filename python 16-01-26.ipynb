Print all elements in a list 

```python
list1 = [10, 11, 12, 14, 15, 13, 16, 17, 18, 20, 19]

for i in list1:
    print(i,end=" ")
```

Find sum of elements in a list 

```python
sum = 0
for item in list1:
    sum += item
print(sum)
```

Find largest number in a list 

```python
for item in range(len(list1)-1):
    if list1[item] > list1[item +1]:
        largest = list1[item]

print(largest)
```

Count even numbers in a list 

```python
count = 0
for i in range(len(list1)):
    if list1[i] % 2 == 0:
        count +=1
print(count)
```

Print list elements in reverse order

```python
list1 = [10, 11, 12, 14, 15, 13, 16, 17, 18, 20, 19]

for item in range(len(list1)-1,-1,-1):
    print(list1[item],end=" ")
```
