## The + operator cannot concatenate a string (item) directly to a list (cart). 



````python
max_items = 5
cart = []
count = 0

while count < max_items:
    
    item = input("Add items to the cart (or enter done)").strip()
    
    if item == "done" or count == 5:
        break
    
    cart = cart + [item]
    print ("You have added {} to the cart".format(cart))
    
    count +=1
    
print("Cart has {} items".format(cart))
````

### Program 11: Login System with Lockout 
#### Problem: Create login system: 
#### • 3 attempts allowed 
#### • Lock account after 3 wrong attempts 
#### • Break on successful login 
 
correct_username = "admin" 
correct_password = "admin123" 
### Your task: Implement complete login system

````python
correct_username = "admin"
correct_password = "admin123" 
max_attempts = 3
attempt = 0

while attempt < max_attempts:
    
    username = input("Enter Username :")
    password = input("Enter password :")
    
    if username == correct_username:
        
        if password == correct_password:
            
            print("Login Successfull!")
            break
        
        else:
            print("Wrong Password!")
            print("{} attemps remaining".format(max_attempts - attempt))
            
    else:
        print("Invalid Username and password")
        
    attempt += 1
    print("{} attempt".format(attempt))

    
if attempt == 3:
    
    print("Account Locked !!!!")

````

## Program 12: Nested Loop - Pattern Breaking 
#### Problem: Print pattern but stop when sum of i+j equals 5. 

#### Your task: 
#### Print (i, j) pairs 
#### Use break when i + j == 5 
#### i ranges from 1-5, j ranges from 1-5 
``` 
**Expected Output:** 
``` 
(1, 1) 

(1, 2) 

(1, 3) 

(1, 4) 

Sum reached 5! Stopping. 

````python
import sys
for i in range(1,6):
    for j in range(1,6):
        if i == 1:
            print(i,j)
            if i + j == 5:
                print("Sum reached {}! Stopping.".format(i+j))
                break
        
````

````python
import sys

for i in range(1, 6):
    for j in range(1, 6):
        if i == 1:
            print(f"({i}, {j})")
            if i + j == 5:
                print(f"Sum reached {i+j}! Stopping.")
                sys.exit()
````

## Program 17: Print Multiplication Table (Skip Even Results) 
#### Problem: Print 5's table from 1-10, but skip even results using continue. 

number = 5 
#### Your task: Skip even products (10, 20, 30, etc.) 
``` 
**Expected Output:** 
```
5 x 1 = 5 

5 x 3 = 15 

5 x 5 = 25 

5 x 7 = 35 

5 x 9 = 45

````python
table_no = int(input("Enter the table number :"))
for n in range(1,11):
    
    res = table_no * n 
    
    if res % 2 == 0:
        
        continue

    else:
        
        print("{} * {} = {}".format(table_no,n,res))
````
