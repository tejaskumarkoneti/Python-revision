## Convert to uppercase - without built in

```python
res = ""

text ="python"

for ch in text:
    
    if ord(ch) >= ord("a") and ord(ch) <= ord("z"):
        
        res_int = ord(ch) - 32
        
        res_ch = chr(res_int)
        
        res = res + res_ch

print(res)
```

## lower() – Convert to lowercase 

```python
email = "NAGA@GMAIL.COM" 
email.lower() 
```

## Convert to lowercase without built in (not expected output)

```python
email = "NAGA@GMAIL.COM" 

res = ""

for ch in email:
    
        
    if ord(ch) >= ord("A") and ord(ch) <= ord("Z"):
        
        res_int = ord(ch) + 32
        
        res_ch = chr(res_int)
        
        res += res_ch
    
    else: 
    
        res += ch

print(res)
```

```python
##title() – First letter of every word capital 
name = "naga balla" 
name.title() 
#####   Output: "Naga Balla" 
#####    Use case: Names, headings
```

```python
##capitalize() – Only first character capital 
msg = "hello world" 
msg.capitalize() 
#   Output: "Hello world" 
#    Use case: Sentence formatting
```

```python
#swapcase() – Upper ↔ Lower 
text = "PyThOn" 
text.swapcase() 
#   Output: "pYtHoN" 
#    Use case: Text transformations, fun effects 
```

```python
 #    Trimming Functions (Very Common) 
 #strip() – Remove spaces (both sides) 
name = "  Naga  " 
name.strip() 
 #  Output: "Naga" 
 #   Use case: User input cleaning
```

```python
# lstrip() – Remove left spaces 
text = "   Hello" 
text.lstrip() 
#   Output: "Hello"
```

```python
# rstrip() – Remove right spaces 
text = "Hello   " 
text.rstrip() 
## Output: "Hello
```

```python
#    Replace & Modify (Creates NEW string) 
 #replace(old, new) 
msg = "I like Java" 
msg.replace("Java", "Python") 
  # Output: "I like Python" 
   # Use case: Updating text, masking words
```

```python
 #split() – String → List 
 
text = "Python is easy" 

text.split() 

 #  Output: ["Python", "is", "easy"] 
 
  #  Use case: Breaking sentences, CSV data 
```

```python
# join() – List → String 
words = ["Python", "is", "easy"] 
" ".join(words) 
 #  Output: "Python is easy"
```

```python
#Searching & Checking 
#find() – Returns index / -1 
text = "Python" 
text.find("o") 
#Output: 4 
text.find("z") 
#Output: -1
```

```python
#index() – Returns index / error 
text.index("o")   # 4 
text.index("z")   #   
#Error
```

```python
# count() – Count occurrences 
text = "banana" 
text.count("a") 
# Output: 3
```

```python
# startswith() 
url = "https://google.com" 
url.startswith("https") 
#Output: True 
```

```python
# endswith() 
file = "data.csv" 
file.endswith(".csv") 
# Output: True
```

```python
# Validation Functions (Forms, Login, OTP) 
# isdigit() 
"12345".isdigit()   # True 
"12a45".isdigit()   # False 
# OTP, phone number
```

```python
#isalpha() 
"Python".isalpha()   # True 
"Python3".isalpha()  # False 
```

```python
# isalnum() 
"User123".isalnum()  # True 
```

```python
# isspace() 
"  ".isspace()   # True 
```

```python
# islower() / isupper() 
"python".islower()   # True 
"PYTHON".isupper()   # True
```

```python
#Slicing Strings (Very Powerful) 
#Format 
# string[start : end : step] 
# First 3 characters 
text = "Python" 
text[0:3] 
#➡ "Pyt" 
#Reverse string 
text[::-1] 
#➡ "nohtyP" 
#Skip characters 
text[::2] 
#➡ "Pto" 
#Important: 
#• Slicing never changes original string 
#• Always creates a new string
```
