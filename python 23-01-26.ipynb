## Print numbers from 1 to 10

```python
i = 1
n = 10
while i <= n:
    print(i,end=" ")
    i+=1
```

## Print even numbers from 1 to 50

```python
i = 1
n = 50
while i <= n:
    if i % 2 == 0:
        print(i,end=" ")
    i+=1
```

## Print odd numbers from 1 to 50

```python
iter = 1
n = 50
while iter <= n:
    if iter % 2 != 0:
        print(iter,end=" ")
    iter +=1
        
```

## Find the sum of first N natural numbers
Input : n = 8
Output : 36

```python
sum = 0
n = 8
for i in range(n,0,-1):
    sum += i
print(sum)
```

## Find the factorial of a given number
Input : n = 5
Output : 120

```python
fact = 1
for i in range(5,0,-1):
    fact *= i
print(fact)
```

## Reverse a given number
Input :
number = 124
Output
421

```python
number = int(input())
temp = number
rev = ""
while number > 0:
    last = number % 10
    rev = rev + str(last)
    number = number // 10
print("Reverse of {} is {}".format(temp,rev))
```

##  Count the number of digits in a number 
Input : 4389
Output : 4

```python
count = 0
n = 438
while n > 0:
    print("inside while loop",n)
    count +=1
    n = n // 10
print(count)
```

## Check whether a number is a palindrome
Input : "12221"
Output : "Palindrome"

Input : "12211"
Output : " Not Palindrome"

```python
rev=""
num = 1222
dummy = num

while num > 0:
    last = num % 10
    rev += str(last)
    num //= 10
    
if int(rev) == dummy:
    print("Palindrome")
else:
    print("Not Palindrome")    
```

## Check whether a number is an Armstrong number

```python
n = int(input("Enter the number to check it is Armstrong number or not :"))
res = 0
temp = n
power = 3
while n > 0:
    last = n % 10
    res = res + ((last) ** power )
    n = n // 10
if res == temp:
    print("Amstrong No")
else:
    print("Not Amstrong No")
    
```
