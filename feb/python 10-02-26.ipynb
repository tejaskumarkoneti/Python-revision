• One function to calculate ticket price  
• Another function to calculate GST

```python
def book_ticket(seat_type,ticket):
    
    price = 200
    if seat_type == "premium":
        price += 150
    
    return price * ticket

total_price = book_ticket("premium",2)
print(total_price)

def gst(amount):
    
    return amount + (( amount * 18) / 100)
    
g = gst(total_price)
print(g)
```
