## Generate Fibonacci series up to Nth terms

Input: 7
Output: 0 1 1 2 3 5 8

Input: 9
Output: 0 1 1 2 3 5 8 13 21

```python
# prints fibonacci series upto n terms
x = 0
y = 1
n = 7
while x <= n:
    print(x,end=" ")
    x , y = y , x + y
```

Generate Fibonacci series up to Nth terms

Input: 7
Output: 0 1 1 2 3 5 8

```python
# prints first 8 fibonacci series  
x = 0
y = 1
n = 1
while n <= 8:
    print(x,end=" ")
    x ,y = y,x + y
    n +=1
```

Generate Fibonacci series up to Nth terms

Input: 8
Output: 0 1 1 2 3 5 8  13

# we will use count variable to count for handling N terms

```python
x = 0
y = 1
n = int(input("Enter the input :"))
print(f"Fibonacci Series  {n} terms are :",end=" ")
count = 0
while count < n:
    print(x,end=" ")
    x ,y = y,x+y
    count +=1
```
