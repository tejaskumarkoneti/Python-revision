Print all prime numbers between 1 and 50 

```python
def prime(n):
    if n <=1:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False    
    return True

for num in range(1,51):
    if prime(num):
        print(num,end=" ")
```

```python
def prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    i = 2
    while i < n:
        if n % i == 0:
            return False
        i += 1
        return True
for num in range(1,51):
    if prime(num):
        print(num,end=" ")
```
