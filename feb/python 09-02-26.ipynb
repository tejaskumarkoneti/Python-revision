### 1. Create a set from a list [10, 20, 20, 30, 40, 40, 50] and print only unique values.

```python
li = [10, 20, 20, 30, 40, 40, 50]

s = set(li)

print(s)
```

### 2. Write a program to check whether a given element exists in a set.

```python
if 50 in s:
    print("Yes")
```

### 3. Add multiple elements {5, 6, 7} to an existing set {1, 2, 3, 4}.

```python
set1 = set([1,2,3,4])

set1.update([5,6,7])

print(set1)
```

### 4. Remove an element from a set. Handle the case where the element does not exist.

```python
try:
    set1.remove(9)
except KeyError:
    print("Invalid")
else:
    print(set1)
finally:
    print("Execution Passed")
```

### 5. Find the union, intersection, and difference of two sets:
- A = {1, 2, 3, 4}  
- B = {3, 4, 5, 6}

```python
A = {1, 2, 3, 4}  
B = {3, 4, 5, 6}

print(A | B)

print(A & B)

print(A - B)
```

### 6. Write a program to find common elements between two lists using sets.

```python
l1 = [1,2,3]
l2 = [3,4,5]

print(set(l1) & set(l2))
```

### 7. Check whether one set is a subset of another set.

```python
s1 = set([1,2,3])
s2 = set([1])


print(s1.issubset(s2))
```

### 8. Remove all duplicate values from a list using a set and convert the result back to a list.

```python
li = [10, 20, 20, 30, 40, 40, 50]

l = list(set(li))

print(l)
```

### 9. Write a program to find elements that are present in only one of two sets.  

### non - common elements in both the sets

```python
s1 = set([1,2,3])
s2 = set([3,4,5])

print(s1 - s2)
print(s2 - s1)

res = (s1 - s2) | (s2 - s1)
print(res)

print(s1 ^ s2)
```
