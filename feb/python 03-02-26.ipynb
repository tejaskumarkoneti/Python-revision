## upper() – Convert to uppercase 

```python
text = "python" 
```

```python
print(text.upper())
```

## Convert to uppercase - without built in

```python
res = ""

text ="python"

for ch in text:
    
    if ord(ch) >= ord("a") and ord(ch) <= ord("z"):
        
        res_int = ord(ch) - 32
        
        res_ch = chr(res_int)
        
        res = res + res_ch

print(res)
```

## lower() – Convert to lowercase 

```python
email = "NAGA@GMAIL.COM" 
email.lower() 
```

## Convert to lowercase without built in (not expected output)

```python
email = "NAGA@GMAIL.COM" 

res = ""

for ch in email:
    
        
    if ord(ch) >= ord("A") and ord(ch) <= ord("Z"):
        
        res_int = ord(ch) + 32
        
        res_ch = chr(res_int)
        
        res += res_ch
    
    else: 
    
        res += ch

print(res)
```

```python
##title() – First letter of every word capital 
name = "naga balla" 
name.title() 
#####   Output: "Naga Balla" 
#####    Use case: Names, headings
```

```python
##capitalize() – Only first character capital 
msg = "hello world" 
msg.capitalize() 
#   Output: "Hello world" 
#    Use case: Sentence formatting
```

```python
#swapcase() – Upper ↔ Lower 
text = "PyThOn" 
text.swapcase() 
#   Output: "pYtHoN" 
#    Use case: Text transformations, fun effects 
```
