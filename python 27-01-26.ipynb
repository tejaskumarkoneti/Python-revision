## Display a menu repeatedly until the user chooses Exit


```python
user = input("Type Menu :")
while user != "Exit":
    print("Menu Items are :")
    print("1.Dry Fruit")
    print("2.Bread ")
    print("3.Milk Products")
    user =input("Choose Menu or Exit")
    if user =="Exit":
        print("user exited")
        break
```

## Display a menu repeatedly until the user chooses Exit(updated)

```python
while True:
    print("\n Menu Items are :")
    print("1.Dry Fruit")
    print("2.Bread ")
    print("3.Milk Products")
    user =input("Choose Menu or Exit")
    if user =="Exit" or user == "exit":
        print("user exited")
        break
```

## Read numbers from the user until they enter 0 and find the sum 

```python
num = int(input())
print(num)
sum = 0
while num != 0:
    sum += num
print(sum)
```

## Read numbers from the user until they enter 0 and find the sum (updated)


```python
sum = 0
while True:
    num = int(input())
    print("User entered {}".format(num))
    sum += num
    if num == 0:
        print(sum)
        break
```
