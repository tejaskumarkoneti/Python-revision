Check whether a number is prime 

```python
num = 12
is_prime = False

while num >= 2:
    if num % 1 == 0 and num % num == 0:
        is_prime = True
        break

if is_prime == True:
    print("prime")
else:
    print("not prime")
```

# here 12 is not a prime number but it is printing it as a prime
