##  Find the largest number entered until the user types “stop”   

````python
user = None
li = []

while user != "stop":
    user = int(input(" Enter the no "))
    li += [user]

for i in range(len(li)-1):
    if li[i] > li[i+1]:
        largest = li[i]
print("Lagest number is ",largest)
````

## Program 1: Find First Negative Number 
Problem: Find the first negative number in a list and stop searching. 

numbers = [10, 20, -5, 30, -8, 40] 
### Your task: Use break to stop when first negative is found 
``` 
**Expected Output:
``` 
First negative number found: -5

````python
numbers = [10,20,-5,30,-8,40]

for number in numbers:
    if number < 0:
        print(f"First Negative no is {number} found")
        break
````

## Program 2: Print Only Positive Numbers 
Problem: Print only positive numbers, skip negative ones using continue. 


numbers = [5, -3, 8, -1, 12, -7, 15] 
# Your task: Use continue to skip negative numbers 
``` 
**Expected Output:** 
``` 
5 
8 
12 
15

````python
numbers = [5,-3,8,-1,12,-7,15]

for number in numbers:
    
    if number > 0:
        
        print(number,end=" ")
    else:
        
        continue
````
